```python
temp_readings =[12,15,30,47,47,60,24,57,20,10,39,45]
normal_temp=0
high_temp=0
for i in temp_readings:
  if i>15 and i<30:
    normal_temp+=1
  elif i>30:
    high_temp+=1
print("normal temperature",normal_temp,"\nhigh temperature",high_temp)

```

#Task 5

```python
speeds = [60, 85, 72, 90, 110, 78, 95, 65]
violation_count = 0
total_vehicles = len(speeds)
for speed in speeds:
    if speed > 80:
        violation_count += 1
violation_percentage = (violation_count / total_vehicles) * 100
print("Total vehicles:", total_vehicles)
print("Overspeeding vehicles:", violation_count)
print("Violation percentage:", violation_percentage, "%")
```
