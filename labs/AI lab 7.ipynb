Task no 1

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# Load datasets
df_csv = pd.read_csv("applicants.csv")
df_json = pd.read_json("applicants_extra.json")

# Inspect CSV
print("CSV HEAD")
print(df_csv.head())

print("\nCSV TYPES")
print(df_csv.dtypes)

print("\nCSV MISSING")
print(df_csv.isnull().sum())

print("\nCSV SHAPE")
print(df_csv.shape)

# Inspect JSON
print("\nJSON HEAD")
print(df_json.head())

print("\nJSON TYPES")
print(df_json.dtypes)

print("\nJSON MISSING")
print(df_json.isnull().sum())

print("\nJSON SHAPE")
print(df_json.shape)
```

Task no 2

```python
# Merge both datasets

df = pd.merge(df_csv, df_json, on="id", how="left")

# Display merged dataframe
print(df)

# Shape
print("\nShape:")
print(df.shape)
```

Task no 3


```python
# ============================================
# AGE -> fill with median
# ============================================

df["age"] = df["age"].fillna(df["age"].median())

# ============================================
# EXPERIENCE -> fill with 0
# ============================================

df["experience"] = df["experience"].fillna(0)

# ============================================
# EXPECTED SALARY
# Replace ? with NaN first
# ============================================

df["expected_salary"] = df["expected_salary"].replace("?", np.nan)

# Convert to numeric
df["expected_salary"] = pd.to_numeric(df["expected_salary"])

# Fill NaN with median
df["expected_salary"] = df["expected_salary"].fillna(
    df["expected_salary"].median()
)

# ============================================
# EDUCATION
# ============================================

df["education"] = df["education"].fillna("Unknown")

# ============================================
# NAME
# ============================================

df["name"] = df["name"].fillna("Unknown")

# Check missing values again
print(df.isnull().sum())
```

Task no 4

```python
# Replace isl with Islamabad
df["city"] = df["city"].replace("isl", "Islamabad")

# Convert all cities to lowercase
df["city"] = df["city"].str.lower()

# Unique values
print(df["city"].unique())
```

Task no 5

```python
def experience_level(exp):

    if exp <= 1:
        return "junior"

    elif exp <= 5:
        return "mid"

    else:
        return "senior"


df["experience_level"] = df["experience"].apply(experience_level)

print(df[["experience", "experience_level"]])
```

```python
# Columns to normalize
cols = ["age", "experience", "expected_salary"]

# Create scaler
scaler = MinMaxScaler()

# Apply normalization
df[cols] = scaler.fit_transform(df[cols])

# Display normalized columns
print(df[cols])
```

Task no 7

```python
def preprocess_applicants(df):

    # ============================================
    # HANDLE MISSING VALUES
    # ============================================

    df["age"] = df["age"].fillna(df["age"].median())

    df["experience"] = df["experience"].fillna(0)

    df["expected_salary"] = df["expected_salary"].replace("?", np.nan)

    df["expected_salary"] = pd.to_numeric(df["expected_salary"])

    df["expected_salary"] = df["expected_salary"].fillna(
        df["expected_salary"].median()
    )

    df["education"] = df["education"].fillna("Unknown")

    df["name"] = df["name"].fillna("Unknown")

    # ============================================
    # CLEAN CITY COLUMN
    # ============================================

    df["city"] = df["city"].replace("isl", "Islamabad")

    df["city"] = df["city"].str.lower()

    # ============================================
    # FEATURE ENGINEERING
    # ============================================

    def experience_level(exp):

        if exp <= 1:
            return "junior"

        elif exp <= 5:
            return "mid"

        else:
            return "senior"

    df["experience_level"] = df["experience"].apply(experience_level)

    # ============================================
    # NORMALIZATION
    # ============================================

    cols = ["age", "experience", "expected_salary"]

    scaler = MinMaxScaler()

    df[cols] = scaler.fit_transform(df[cols])

    return df
```
