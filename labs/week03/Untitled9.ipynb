```python
def remove_bad_readings(data):
    new_list = []
    i = 0
    while i < len(data):
        if data[i] >= 0 and data[i] <= 100:
            new_list.append(data[i])
        i = i + 1
    return new_list


def calculate_mean(values):
    sum = 0
    n = 0
    for x in values:
        sum = sum + x
        n = n + 1

    if n == 0:
        return 0
    return sum / n


readings = [45, 78, -12, 90, 105, 66, 88]
clean = remove_bad_readings(readings)
avg = calculate_mean(clean)

print(readings)
print(clean)
print(avg)
```

```python
def get_average(marks_list):
    sum_marks = 0
    i = 0
    while i < len(marks_list):
        sum_marks = sum_marks + marks_list[i]
        i = i + 1

    if len(marks_list) == 0:
        return 0
    return sum_marks / len(marks_list)


def check_pass_fail(avg_score):
    if avg_score >= 50:
        return "Passed"
    else:
        return "Failed"


def show_report(student, scores):
    average_score = get_average(scores)
    status = check_pass_fail(average_score)

    print("Name:", student)
    print("Quiz marks:", scores)
    print("Average marks:", round(average_score, 2))
    print("Result:", status)
    print("--------------------")


# Test the program
class_data = {
    "Ali": [65, 70, 80],
    "Sara": [90, 85, 88],
    "Ahmed": [40, 55, 60]
}

student_names = list(class_data.keys())
i = 0
while i < len(student_names):
    current_student = student_names[i]
    current_marks = class_data[current_student]
    show_report(current_student, current_marks)
    i = i + 1
```

```python
class Dataset:
    def __init__(self, values):
        self.data = values

    def get_count(self):
        count = 0
        for value in self.data:
            count = count + 1
        return count

    def get_average(self):
        if self.get_count() == 0:
            return 0

        total = 0
        for number in self.data:
            total = total + number

        avg = total / self.get_count()
        return avg


numbers = [12, 18, 25, 30, 22, 27]

my_data = Dataset(numbers)

print("Data values:", my_data.data)
print("Number of values:", my_data.get_count())
print("Average:", my_data.get_average())
```

```python
def remove_invalid(data_list):
    good = []
    for num in data_list:
        if num >= 0 and num <= 100:
            good.append(num)
    return good


class DataSet:
    def __init__(self, raw):
        self.valid = remove_invalid(raw)

    def count(self):
        return len(self.valid)

    def total(self):
        s = 0
        for v in self.valid:
            s += v
        return s

    def mean(self):
        if self.count() == 0:
            return 0
        return self.total() / self.count()

    def smallest(self):
        if self.count() == 0:
            return 0
        min_val = self.valid[0]
        for x in self.valid:
            if x < min_val:
                min_val = x
        return min_val

    def largest(self):
        if self.count() == 0:
            return 0
        max_val = self.valid[0]
        for x in self.valid:
            if x > max_val:
                max_val = x
        return max_val

    def print_report(self):
        print("Valid values:", self.valid)
        print("Count     :", self.count())
        print("Sum       :", self.total())
        print("Average   :", round(self.mean(), 1))
        print("Min       :", self.smallest())
        print("Max       :", self.largest())
        print("------------------------")



numbers = [10, 55, -3, 78, 120, 66, 92]
my_data = DataSet(numbers)

print("Input data:", numbers)
my_data.print_report()
```
