Task no 1

```python
import requests
from bs4 import BeautifulSoup

url = "http://quotes.toscrape.com/"

res = requests.get(url)
soup = BeautifulSoup(res.text, "html.parser")

quotes_data = []

quotes = soup.find_all("div", class_="quote")

for q in quotes:
    text = q.find("span", class_="text").text
    author = q.find("small", class_="author").text

    quotes_data.append({
        "quote": text,
        "author": author
    })

print(quotes_data)
```

Task no 2

```python
all_quotes = []

for i in range(1, 4):
    url = f"http://quotes.toscrape.com/page/{i}/"
    res = requests.get(url)
    soup = BeautifulSoup(res.text, "html.parser")

    quotes = soup.find_all("div", class_="quote")

    for q in quotes:
        text = q.find("span", class_="text").text
        author = q.find("small", class_="author").text

        all_quotes.append({
            "quote": text,
            "author": author
        })

print(len(all_quotes))

for q in all_quotes[:5]:
    print(q)
```

Task no 3

```python
import sqlite3

conn = sqlite3.connect("quotes.db")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS quotes (
    id INTEGER PRIMARY KEY,
    quote TEXT,
    author TEXT
)
""")

print("Table Created")

conn.close()
```

Task no 4

```python
conn = sqlite3.connect("quotes.db")
cursor = conn.cursor()

data = [(q["quote"], q["author"]) for q in all_quotes]

cursor.executemany("INSERT INTO quotes (quote, author) VALUES (?, ?)", data)

conn.commit()
conn.close()
```

Task no 6

```python
conn = sqlite3.connect("quotes.db")
cursor = conn.cursor()

author_name = "Albert Einstein"

cursor.execute("SELECT * FROM quotes WHERE author = ?", (author_name,))
print(cursor.fetchall())

cursor.execute("SELECT author, COUNT(*) FROM quotes GROUP BY author")
print(cursor.fetchall())

cursor.execute("SELECT author, COUNT(*) as c FROM quotes GROUP BY author ORDER BY c DESC")
print(cursor.fetchall())

conn.close()
```
