Part A – Dataset Preparation

Task no 1

```python
# Import libraries
from sklearn.datasets import load_wine
import pandas as pd

# Load dataset
wine = load_wine()

# Features and target
X = wine.data
y = wine.target

# Convert to DataFrame
df = pd.DataFrame(X, columns=wine.feature_names)

# Add target column
df["target"] = y

# Display first 5 rows
print("First 5 Rows:")
print(df.head())

# Display shape
print("\nDataset Shape:")
print(df.shape)
```

Task no 2

```python
from sklearn.model_selection import train_test_split

# Features and target
X = df.drop("target", axis=1)
y = df["target"]

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.30,
    random_state=42
)

# Print shapes
print("Training Shape:", X_train.shape)
print("Testing Shape:", X_test.shape)
```

Part B – Baseline Mode

Task no 3

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Create model
log_model = LogisticRegression(max_iter=5000)

# Train model
log_model.fit(X_train, y_train)

# Predict
predictions = log_model.predict(X_test)

# Accuracy
baseline_accuracy = accuracy_score(y_test, predictions)

print("Baseline Accuracy:")
print(baseline_accuracy)
```

Part C – Cross-Validation

Task no 4

```python
from sklearn.model_selection import cross_val_score

# Logistic Regression model
model = LogisticRegression(max_iter=5000)

# 5-fold cross-validation
scores = cross_val_score(model, X, y, cv=5)

# Print scores
print("Cross-Validation Scores:")
print(scores)

# Mean accuracy
mean_accuracy = scores.mean()

print("\nMean Accuracy:")
print(mean_accuracy)
```

Task no 5

```python
print("Baseline Accuracy:", baseline_accuracy)
print("Cross-Validation Mean Accuracy:", mean_accuracy)

# Comparison
if mean_accuracy > baseline_accuracy:
    print("Cross-validation performed better.")
else:
    print("Baseline train-test split performed better.")
```

Part D – Hyperparameter Tuning

Task no 6

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Different k values
k_values = [3, 5, 7, 9]

# Store results
results = {}

for k in k_values:

    # Create model
    knn = KNeighborsClassifier(n_neighbors=k)

    # Train model
    knn.fit(X_train, y_train)

    # Predict
    pred = knn.predict(X_test)

    # Accuracy
    acc = accuracy_score(y_test, pred)

    # Store result
    results[k] = acc

    print(f"k={k} Accuracy={acc}")

# Best k
best_k = max(results, key=results.get)

print("\nBest k value:", best_k)
```

Task no 7

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier

# Parameter grid
param_grid = {
    "n_neighbors": [3, 5, 7, 9]
}

# KNN model
knn = KNeighborsClassifier()

# GridSearchCV
grid = GridSearchCV(
    knn,
    param_grid,
    cv=5
)

# Train
grid.fit(X_train, y_train)

# Best parameters
print("Best Parameters:")
print(grid.best_params_)

# Best score
print("\nBest Cross-Validation Score:")
print(grid.best_score_)
```

Part E – Model Comparison

Task no 8

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score

# Models dictionary
models = {
    "Logistic Regression": LogisticRegression(max_iter=5000),
    "KNN": KNeighborsClassifier(n_neighbors=best_k),
    "Decision Tree": DecisionTreeClassifier(random_state=42)
}

# Store results
model_results = {}

# Train and evaluate
for name, model in models.items():

    scores = cross_val_score(model, X, y, cv=5)

    mean_score = scores.mean()

    model_results[name] = mean_score

    print(f"{name}: {mean_score}")
```

Part F – Final Model Selection

Task no 9

```python
# Best model name
best_model_name = max(model_results, key=model_results.get)

print("Best Model:")
print(best_model_name)

# Select model
best_model = models[best_model_name]

# Train on training data
best_model.fit(X_train, y_train)

# Predict on test data
final_predictions = best_model.predict(X_test)

# Final accuracy
final_accuracy = accuracy_score(y_test, final_predictions)

print("\nFinal Test Accuracy:")
print(final_accuracy)-+
```
