```python
readings = [15, 45, 82, 67, 19, 90, 55, 23]

alerts = 0

for i in readings:
    if i < 20 or i > 80:
        alerts += 1

percentage = (alerts / len(readings)) * 100

print("Total alerts =", alerts)
print("Alert percentage =", percentage, "%")
```
